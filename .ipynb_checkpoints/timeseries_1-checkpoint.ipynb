## ARIMA: autoregressive integrated moving average: p,d,q acounts for  trend, noise and seasonality 

#### Creat parameter combinations

```python
# create different combination of paarmeters
# p: trend antoregression order, d: trend difference order, q: trend moving average order
p=d=q=range(0,2)
pdq=list(itertools.product(p,d,q))
pdq
```

```python
# P: seasonal antoregression order, D: seasonal difference order, Q: seasonal moving average order, m: The number of time steps for a single seasonal period.
seasonal_pdq=[(x[0], x[1], x[2],12) for x in list (itertools.product(p,d,q))] #m=12 for monthly data
seasonal_pdq
```

```python
print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

```python
seasonal_pdq
```

```python
pdq
```

```python
# using grid search to find the best combination of parameters
parameters=[]
for param in pdq:
    print('pdq:', param)
    for param_seasonal in seasonal_pdq:
        print('param_seasonal:', param_seasonal)
        # The try block lets you test a block of code for errors. and except block handle the error
        try:
            # tsa: time series analysis
            # Seasonal AutoRegressive Integrated Moving Average with eXogenous regressors model
            mod=sm.tsa.statespace.SARIMAX(y, 
                                          order=param,
                                          seasonal_order=param_seasonal,
                                          enforce_stationarity=False,
                                          enforce_invertibility=False
                                         )
            # Fit the model via maximum likelihood
            results=mod.fit()
            # Show the summary of results
#             print(results.summary())
            #print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
            parameters.append((param, param_seasonal, results.aic))
            print(parameters)
        except:
            continue
```

```python
print(parameters, type(parameters))
```

```python
np.array(parameters)
```

```python
mod = sm.tsa.statespace.SARIMAX(y,
                                order=(1, 1, 1),
                                seasonal_order=(1, 1, 0, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = mod.fit()
print(results.summary().tables[1])
```

```python
results.aic
```

```python
results.summary()
```

```python
results.plot_diagnostics(figsize=(16,8))
```

```python
# validating forecasts: compare predicted sale to real sales
pred=results.get_prediction(start=pd.to_datetime('2017-01-01'), dynamic=False)
pred_ci=pred.conf_int()
pred_ci
```

```python
pred
```

# Autocorrelation and Partial Autocorrelation https://machinelearningmastery.com/gentle-introduction-autocorrelation-partial-autocorrelation/

#### Autocorrelation and partial autocorrelation plotssummarize the strength of a relationship with an observation in a time series with observations at prior time steps.

```python
furniture.head()
```

```python
furniture.columns
```

```python
fur1=furniture.copy()
fur1=fur1.set_index('ds')
fur1.head()
```

```python
# AutoCorrelation Function (ACF)
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(fur1)
plt.show()
```

```python
# Partial Autocorrelation Function (PACF): remove indirect autocorrelaiton 
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(fur1, lags=50) # limit lag to 50
plt.show()
```

##### https://medium.com/@stallonejacob/time-series-forecast-a-basic-introduction-using-python-414fcb963000#:~:text=Time%20series%20forecasting%20is%20basically,Handling%20Time%20Series%20in%20Pandas.

```python
# another way to do it
# https://medium.com/@stallonejacob/time-series-forecast-a-basic-introduction-using-python-414fcb963000#:~:text=Time%20series%20forecasting%20is%20basically,Handling%20Time%20Series%20in%20Pandas.
from statsmodels.tsa.stattools import acf, pacf
lag_acf=acf(fur1, nlags=20)
lag_pacf=pacf(fur1, nlags=20)
```

```python
# plot acf:the first time where pcaf crosses the uppper CI, here it is 1. So q=1
plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0, linestyle='--', color='grey')
# 95% CI
plt.axhline(y=-1.96/np.sqrt(len(fur1)), linestyle='--', color='grey')
plt.axhline(y=1.96/np.sqrt(len(fur1)), linestyle='--', color='grey')
plt.title('Autocorrelaiton Function')

# plot pacf: the first time where pcaf crosses the uppper CI, here it is 1. So p=1
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0, linestyle='--', color='grey')
# 95% CI
plt.axhline(y=-1.96/np.sqrt(len(fur1)), linestyle='--', color='grey')
plt.axhline(y=1.96/np.sqrt(len(fur1)), linestyle='--', color='grey')
plt.title('Patial Autocorrelaiton Function')
plt.tight_layout() #automatically adjusts subplot params so that the subplot(s) fits in to the figure area.
```
