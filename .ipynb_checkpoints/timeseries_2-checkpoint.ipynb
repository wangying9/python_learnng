```python
#Importing required Libraries
import numpy as np
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf 
from statsmodels.tsa.seasonal import seasonal_decompose 
from pmdarima import auto_arima                        
from sklearn.metrics import mean_squared_error
from statsmodels.tools.eval_measures import rmse
import warnings
warnings.filterwarnings("ignore")
```

### ARIMA Forecast

```python
#Finding the best values for p,q,d,P,Q,D: set the max values for the test
# best combination is SARIMAX(2, 1, 1)x(4, 0, 3, 12)

auto_arima(df['Monthly_Prod'], seasonal=True, m=12,max_p=7, max_d=5,max_q=7, max_P=4, max_D=4,max_Q=4).summary()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
# train test split
train=df[:len(df)-12]
test=df[len(df)-12:]
arima_model=SARIMAX(train['Monthly_Prod'], order=(2,1,1), seasonal_order=(4,0,3,12))
arima_result=arima_model.fit()
arima_result.summary()
```

```python
# Prediction
## ‘levels’ : Predict the levels of the original endogenous variables.
arima_pred=arima_result.predict(start=len(train), end=len(df)-1,typ='levels').rename('ARIMA Predictions')
arima_pred
```

```python
# plot figure to compare predicted and original data
test['Monthly_Prod'].plot(figsize=(16,5), legend=True)
arima_pred.plot(legend=True);
```

```python
# model evaluation
arima_rmse_error=rmse(test['Monthly_Prod'],arima_pred)
arima_mse_error=arima_rmse_error**2
mean_value=df['Monthly_Prod'].mean()
print(f'MSE Error: {arima_mse_error}\nRMSE Error: {arima_rmse_error}\nMean: {mean_value}')
```

```python
test['ARIMA_Predictions'] = arima_pred
test
```

### Prophet

```python
df_pr=df.copy()
df_pr.head()
```

```python
#change index and col name
df_pr=df_pr.reset_index()
print(df_pr.head())
df_pr.columns=['ds','y']
df_pr.head()
```

```python
# split train and test data
train_data_pr = df_pr.iloc[:len(df)-12]
test_data_pr = df_pr.iloc[len(df)-12:]
```

```python
# model fit
from fbprophet import Prophet
m=Prophet()
m.fit(train_data_pr)
future=m.make_future_dataframe(periods=12, freq='MS')
prophet_pred=m.predict(future)
prophet_pred.tail()
```

```python
# the last 12 month
prophet_pred[-12:]
```

```python
#select the last 12 months
prophet_pred=pd.DataFrame({'Date': prophet_pred[-12:]['ds'], 'Pred': prophet_pred[-12:]['yhat']})

# reset index
prophet_pred=prophet_pred.set_index('Date')
prophet_pred.index.freq='MS'
prophet_pred
```

```python
# plot
test["Prophet_Predictions"]=prophet_pred['Pred'].values
test
```

```python
import seaborn as sns
plt.figure(figsize=(16,5))
ax=sns.lineplot(x=test.index, y=test['Monthly_Prod'])
sns.lineplot(x=test.index, y=test['Prophet_Predictions']);
```

```python
# evaluation 
prophet_rmse_error = rmse(test['Monthly_Prod'], test["Prophet_Predictions"])
prophet_mse_error = prophet_rmse_error**2
prophet_mse_error1 = mean_squared_error(test['Monthly_Prod'], test["Prophet_Predictions"])
mean_value = df['Monthly_Prod'].mean()

print(f'MSE Error: {prophet_mse_error}\nRMSE Error: {prophet_rmse_error}\nMean: {mean_value}\nMSE Error1: {prophet_mse_error1}')
```

```python
# comparison between two methods
## build dataframe
rmse_errors = [arima_rmse_error, prophet_rmse_error]
mse_errors = [arima_mse_error, prophet_mse_error]
errors = pd.DataFrame({"Models" : ["ARIMA", "Prophet"],"RMSE Errors" : rmse_errors, "MSE Errors" : mse_errors})
errors
```

```python
# plot
plt.figure(figsize=(16,9))
plt.plot_date(test.index, test["Monthly_Prod"], linestyle="-")
plt.plot_date(test.index, test["ARIMA_Predictions"], linestyle="-.")
plt.plot_date(test.index, test["Prophet_Predictions"], linestyle=":")
plt.legend()
plt.show()
```
